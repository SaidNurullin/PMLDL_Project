```python
from music21 import converter, note
import matplotlib.pyplot as plt
from collections import Counter

def get_note_distribution(midi_file):
    """
    Extracts note pitches from a MIDI file.
    """
    score = converter.parse(midi_file)
    notes = []
    for element in score.flat.notes:
        if isinstance(element, note.Note):
            notes.append(str(element.pitch))
    return notes

# Analyzing note distribution
generated_notes = get_note_distribution('generated_music.mid')

# Plotting note distribution
generated_note_counts = Counter(generated_notes)
plt.bar(generated_note_counts.keys(), generated_note_counts.values())
plt.title('Note Distribution in Generated Music')
plt.xlabel('Notes')
plt.ylabel('Frequency')
plt.show()
```

```python
from music21 import converter
import matplotlib.pyplot as plt

def get_rhythm_distribution(midi_file):
    """
    Extracts rhythmic values from a MIDI file.
    """
    score = converter.parse(midi_file)
    rhythms = []
    for element in score.flat.notesAndRests:
        rhythms.append(float(element.quarterLength))
    return rhythms

# Analyzing rhythm distribution
generated_rhythms = get_rhythm_distribution('generated_music.mid')

# Plotting rhythm distribution
plt.hist(generated_rhythms, bins=10, alpha=0.7)
plt.title('Rhythm Distribution in Generated Music')
plt.xlabel('Rhythmic Value (in Quarter Length)')
plt.ylabel('Frequency')
plt.show()
```

```python
from music21 import converter

def get_key_signature(midi_file):
    """
    Analyzes and returns the key signature of a MIDI file.
    """
    score = converter.parse(midi_file)
    key = score.analyze('key')
    return key

# Analyzing key signature
generated_key = get_key_signature('generated_music.mid')
print(f"Generated Music Key: {generated_key}")
```

```python
from pretty_midi import PrettyMIDI
import numpy as np

def visualize_piano_roll(midi_file):
    midi_data = PrettyMIDI(midi_file)
    piano_roll = midi_data.get_piano_roll()

    plt.imshow(piano_roll, aspect='auto', cmap='gray_r')
    plt.title('Piano Roll Visualization')
    plt.xlabel('Time')
    plt.ylabel('Pitch')
    plt.show()

# Визуализация пианорола
visualize_piano_roll('generated_music.mid')
```

```python
from music21 import roman, converter

def get_chord_progressions(midi_file):
    """
    Extracts chord progressions from a MIDI file in Roman numeral notation.
    """
    score = converter.parse(midi_file)
    chord_progressions = []
    
    for part in score.parts:
        for chord in part.chordify().recurse().getElementsByClass('Chord'):
            roman_numeral = roman.romanNumeralFromChord(chord, score.analyze('key'))
            chord_progressions.append(str(roman_numeral))
    
    return chord_progressions

# Analyzing chord progressions
generated_chords = get_chord_progressions('generated_music.mid')
print(f"Chord Progressions in Generated Music: {generated_chords}")
```
